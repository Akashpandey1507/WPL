
# Importing Important Liabraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#  Reading Data

```python
path = r'E:\CSV_Files\Project\Wpl'
```

```python
df1 = pd.read_csv(fr'{path}\auction_2023.csv')
df2 = pd.read_csv(fr'{path}\players_info_2023.csv')
df1
```

```python
df2
```

# Merging the Data

```python
df = df1.merge(df2,how='inner',on='Player')
df
```

# Data Cleaning

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df = df.drop(columns=['Price (₹)','Price','National Side_y','Full Name','Team_y','Born','Teams Played'])
```

```python
df.head()
```

```python
df = df.rename(columns={
    'National Side_x' : 'Country',
    'Team_x' : 'Team',
})
```

```python
df['Age'] = df['Age'].str[0:2]
```

```python
df.info()
```

```python
df = df.astype({
    'Age' : 'int64'
})
```

```python
df
```

```python
df.describe().T
```

```python
plt.figure(figsize=(15,5))
plt.title('County wise Player in WPL 2023')

sns.countplot(data = df,x=df['Country'])
plt.show()
```

```python
exp_franchise = df.groupby('Team')['Price (US$)'].sum().reset_index()
exp_franchise
```

```python
plt.figure(figsize=(15,5))
plt.title('Total Expenditure by Franchise to purchase PLayer for Playing')

sns.barplot(data = exp_franchise,x='Team',y='Price (US$)')
plt.show()
```
